## Uniform distribution

```python
#A uniform distribution is the one for which the values have the same probability
import numpy as np
import matplotlib.pyplot as plt
#Where we import the numpy for doing the computations, and matplotlib for the graphical representations

a = 1
b = 100
n = 200
#The values we are going to make for the uniform distribution

data = np.random.uniform(a,b,n)
#Where a is the initial value it can take, the b is the end of the interval. N is the number of values we get.

%matplotlib inline
plt.hist(data)
#Plotting an histogram
```

## Normal distribution

```python
#We are going to use randn to create the normal distribution (the standard one has mean 0 and sd of 1)
n=100

data = np.random.randn(n)

x = range(1,101)
#X must have same first dimensions as in n

plt.plot(x, data)
#We will plot the different values we have obtained
```

```python
#Creating a non-standard normal distribution
mu = 5.5
sd = 2.5
z_values = np.random.randn(n)
#Will have a mean of 5.5, and a standard deviation of 5.5

data = mu + sd * z_values

%matplotlib inline
plt.hist(data)
```

```python
#How randn works
np.random.randn(2,5)

#First number is going to create a the number of normal distributions
#The second one is how much values each normal distribution will have
```
