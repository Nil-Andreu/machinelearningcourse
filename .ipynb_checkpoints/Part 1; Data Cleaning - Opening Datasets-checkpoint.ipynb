## Reading a database by rows

### Importing the dataset

```python
#We can open open a file by parts (line of code by line of code). First we will start preparing the dataset.

#We can make a mainpath for the directory were I will have all the datasets
mainpath = r"C:\Users\Msi\Documents\YO\Programación\Machine Learning\datasets"

#And then having a specific path for the dataset I am looking for
filepath = r"\customer_churn_model\customer_churn_model.txt"

#Then we sum up both to get the full path
fullpath = mainpath + filepath

#And use the function of open, which will allow us to open it by parts
data2 = open(fullpath)
```

```python
#If we look to th efile, we can see that it is a Text File divided by commas
#So we will start extracting the columns

cols = data2.readline().strip().split(",")
#Where the readline what makes is to read the first line of the file
#The strip method what does is to eliminate the white spaces
#The split method what makes is to divide the colums by the commas
```

```python
#So now, we will have an array with the values of the columns
cols
```

### Making an script to read the database

```python
#Now, we are going to make a basic script to pass through the rows of the dataset
counter = 0
main_dict = {}
#Where in this dictionary we are going to have the for each key (which is the name of the columns), 
#an empty set of values that we will later put

for col in cols:
    main_dict[col] = []
```

```python
#We can see 
main_dict
```

```python
#Now, we make the script itself
for line in data2:
    #Where we are iterating through each line of the dataset
    
    values = line.strip().split(",")
    #We are obtaining all the values that are in the dataset
    
    for i in range(len(cols)):
        #What we are doing is an iteration for the lenght of each column 
        # (this way we can pass the values to the one corresponding)
        
        main_dict[cols[i]].append(values[i])
        #We append each value of the line to the position that corresponds in the column (with columns[i])
        #So this way, if e.g. we have 10 columns, the value number 4 will be positioned there
    counter += 1
    
print(f"We are having {counter} lines") 
#We pass a formated string with the counter of lines to make sure it run properily
```

```python
#We create now the DataFrame from the dictionary using the function of pandas of DataFrame
dataframe2 = pd.DataFrame(main_dict)

dataframe2.head()
#A DataFrame is how the pandas will be able to read our data. We can use the .head() to ensure all is correct
```
