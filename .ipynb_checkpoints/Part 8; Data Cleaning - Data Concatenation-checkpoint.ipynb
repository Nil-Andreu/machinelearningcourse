## Concatenation

```python
import pandas as pd

#We will use the dataset of wine to do the concatenations
filepath = 'datasets/wine/' #Where the dataset we are going to use is
red_wine = pd.read_csv(filepath + 'winequality-red.csv', sep=";") #As the dataset is separated by ;
white_wine = pd.read_csv(filepath + 'winequality-white.csv', sep=";")

red_wine.head()
```

```python
#Make sure the wine dataset is also imported correctly
white_wine.head()
```

```python
#Now, we make the concatenation with the pandas method of concat
wine_data = pd.concat([red_wine, white_wine], axis=0)

wine_data.head()
#We will see that the first values are the ones of the red wine
```

#### Quick note
In Python, there are two possible axis:
* The 0 axis is to make the concatenation in x
* The 1 axis is the column one

```python
#We could see that if we sum the number of rows of white and red is the same as the total
len(red_wine)+len(white_wine)==len(wine_data)
```

## Distributed data

Sometimes the data that we want to analyse isn't in two or three datasets, but is distributed with hundreds of them. So we want to automate the concatenation of this dataset

```python
#WE are going to use the distributed dataset
filepath = 'datasets/distributed-data/'

data = pd.read_csv(filepath + '001.csv')
data.head() #To see which is the structure of the first file
```

```python
#We can see which is the final value
data.tail()
```

```python
#Now, we are going to iterate to create it
for i in range(2,333): #We make the range of 2 (as the first one is already in data), and 333 as we have 332 datasets
    if i < 10:
        filename = pd.read_csv(filepath + '00' + str(i) + '.csv') 
        #As the name of files have consistency, we need to follow the same structure
    if 10 <= i < 100:
        filename = pd.read_csv(filepath + '0' + str(i) + '.csv')
    if i >= 100:
        filename = pd.read_csv(filepath + str(i) + '.csv')
    
    data = pd.concat([data, filename], axis=0) #We concatenate for the filename each iteration is doing
```
