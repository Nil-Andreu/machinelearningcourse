## Dummy variables

```python
#We can convert dummy variables (the ones that only have true or false), to be able to use it easily
titanic["sex"]
```

```python
#The get_dummies will create as much columns as dummy values it has
dummy_sex = pd.get_dummies(titanic["sex"], prefix="sex")

#Where in the prefix we say which will be the prefix for the columns created: sex_male & sex_female
#This process is known as dummification
```

```python
dummy_sex.head()
```

```python
#We can see that the sex column now is eliminated
titanic.columns.values.tolist()
```

```python
#We need to add the new columns
titanic = pd.concat([titanic, dummy_sex], axis=1)
```
