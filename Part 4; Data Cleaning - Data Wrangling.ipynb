## Substracting columns

```python
#A simple substraction of only one column
account_length = data["Account Length"]
account_length.head()
```

```python
#Substracting a column, we will obtain a series object
type(account_length)
```

```python
#We can make a substraction of a group of columns
desired_columns = ["Account Length", "Phone", "Eve Charge"]
subset = data[desired_columns]
subset.head()
```

```python
#Now when there is multiple columns, what we obtain is a DataFrame
type(subset)
```

```python
#We can do complementaries to obtain the other columns (if for example now we wanted to work with the others)
all_columns_list = data.columns.values.tolist()
sublist = [x for x in all_columns_list if x not in desired_columns]
sublist

#We obtain the columns we do not want
```

```python
#We could obtain the complementaries this way (imagine that now we wanted this ones, and not the desired_columns)
subset = data[sublist]
subset.head()

#This is useful to work as we can use the one which is the smallest, so this way we will be more efficient
```

## Substracting rows

```python
#We can also substract different rows
data[1:20]

#Where the first number is take into consideration, but the last one not
#We could also do for example: data[30:500]
```

```python
#We may want to do a substraction with conditionality
data1 = data[data["Day Mins"] > 300] #For collecting the rows that have grater than 300 minutes in Day Mins
data2 = data[data["Day Mins"] <= 200] #For the ones having less or equal
data3 = data[data["State"] == "OH"] #For the ones that are in Ohio only
```

```python
#We could use also double conditionals: AND and OR
data4 = data[(data["Day Mins"] > 300 )&(data["State"] == "OH")] #For the ones that are in Ohio and also have > 300 mins
data5 = data[(data["Day Mins"] > 300 )&(data["State"] == "OH")] #For the ones that are in Ohio or have > 300 mins
```

```python
#We could also filter with the values of the rows itself
data6 = data[data["Day Mins"] > data["Night Mins"]] #Having a greater minuts in day than in night
data6
```
